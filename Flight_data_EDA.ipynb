```python
# number of Flight per month
number_of_Flight_per_month = Flight_Data.month.value_counts()
number_of_Flight_per_month
```

```python
number_of_passenger_per_year = Flight_Data.groupby("year")["passengers"].sum()
number_of_passenger_per_year
```

```python
# number of passengers per month 
number_of_passenger_per_month = Flight_Data.groupby("month")["passengers"].sum()
number_of_passenger_per_month
```

```python
# avg passengers per month 
avg_passengersa_per_month = Flight_Data.groupby("month")["passengers"].mean()
avg_passengersa_per_month
```

```python
#number of flights per year
number_of_flights_per_year = Flight_Data.year.value_counts()
number_of_flights_per_year
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,6))
sns.lineplot(data=Flight_Data,x="year",y="passengers",estimator=sum,ci=False)
plt.title("Number of passengers per year")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(data=Flight_Data,x="month",y="passengers",ci=False,estimator=sum)
plt.title("Number of passengers per month")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(data=Flight_Data,x="month",y="passengers")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=Flight_Data,x="year",y="passengers",estimator=sum,ci=False,hue="month")
plt.title("passengers count over year and month ")
```
